```python
#Example of Bfill
"""
Original:
      A      B
  0   1.0    NaN
  1   NaN    2.0
  2   NaN    NaN
  3   4.0    3.0
  4   5.0    NaN

After bfill:
      A       B
  0   1.0     2.0 #NaN filled with next valid value
  1   4.0     2.0 #NaN filled with next valid value
  2   4.0     3.0 #NaN filled with next valid value
  3   4.0     3.0
  4   5.0     NaN #stayNaN(no next value available)
  """
```

```python
#Example of ffill
"""
Original:
      A
  0   1.0
  1   NaN
  2   3.0
  3   NaN
  4   5.0

After interpolate:
      A
  0   1.0
  1   2.0 #filled between 1 and 3
  2   3.0
  3   4.0 #filled between 3 and 5
  4   5.0
  """
```
